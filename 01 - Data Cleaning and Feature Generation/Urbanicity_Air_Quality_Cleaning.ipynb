## Building features

```python
stateyear_aq.head()
```

```python
# Make a feature-specific DataFrame
aq_feat = stateyear_aq[[
  'State','Year'
]].copy()
```

```python
stateyear_aq.columns
```

```python
# Looping previous years' AQ data into feature columns
for col in stateyear_aq.columns[2:]:
  aq_feat.loc[:,f'Prev {col}'] = stateyear_aq.loc[:,col].copy()

# Add a knowledge year for these "previous AQ" features
aq_feat.loc[:,'Knowledge Year'] = aq_feat.Year + 1

# Make knowledge year the default year index
aq_feat.drop(columns = 'Year', inplace=True)
aq_feat.rename(columns={'Knowledge Year':'Year'}, inplace=True)
```

```python
aq_feat[
  aq_feat.State == 'Alabama'
].head()
```

```python
# 5-year rolling average for each state
aq_5yr_avg = aq_feat.sort_values('Year').groupby('State', as_index=False).rolling(5, min_periods=1).mean().drop(columns='Year')

# Renaming the rolling avg columns
newcolnames = list()
for col in aq_5yr_avg.columns:
  newcolnames.append('5 Year Avg ' + col[5:])
aq_5yr_avg.columns = newcolnames
```

```python
aq_5yr_avg.head()
```

```python
print(f'{len(aq_feat)} and {len(aq_5yr_avg)}')
```

```python
# Concatenating feature DataFrames
aq_features = pd.concat(
  [
    aq_feat.sort_values(['State','Year']).reset_index(),
    aq_5yr_avg.reset_index()
  ], axis=1
)
```

```python
aq_features.head()
```

```python
# Preserving only the features I want
aq_features = aq_features.loc[:,[
  'State', 'Year', 'Prev Days with AQI', 'Prev Good Days',
  'Prev Moderate Days', 'Prev Unhealthy for Sensitive Groups Days',
  'Prev Unhealthy Days', 'Prev Very Unhealthy Days',
  'Prev Hazardous Days', 'Prev Max AQI', 'Prev 90th Percentile AQI',
  'Prev Median AQI', 'Prev Days CO', 'Prev Days NO2', 'Prev Days Ozone',
  'Prev Days PM2.5', 'Prev Days PM10',
  '5 Year Avg Days with AQI', '5 Year Avg Good Days',
  '5 Year Avg Moderate Days',
  '5 Year Avg Unhealthy for Sensitive Groups Days',
  '5 Year Avg Unhealthy Days', '5 Year Avg Very Unhealthy Days',
  '5 Year Avg Hazardous Days', '5 Year Avg Max AQI',
  '5 Year Avg 90th Percentile AQI', '5 Year Avg Median AQI',
  '5 Year Avg Days CO', '5 Year Avg Days NO2', '5 Year Avg Days Ozone',
  '5 Year Avg Days PM2.5', '5 Year Avg Days PM10'
]]
```

```python
# Export to CSV
aq_features.to_csv('../04 - Data/Final Data/aq_features.csv')
```

# Urbanicity

```python
# I had to install this dependency when I got an error.
# But now I have this piece run in the imports cell
#!pip install xlrd==1.2.0

# Testing import on a visually-formatted table
urb = pd.read_excel(
  '../04 - Data/Raw_Data/pop-urban-pct-historical.xls',
  sheet_name='States', skiprows=5
)

# Preview table
urb.head()
```

```python
urb.columns
```

```python
# Cutting unnecessary columns and renaming stuff
urb_small = urb[
  urb['Area Name'] != 'United States'
].copy().iloc[:,[
  1,-2,-1
]]
urb_small.columns = ['State','year_2000','year_2010']

# Checking to see that it worked
urb_small.info()
```

```python
# Defining imputation for urbanicity feature (with minimal data leakage)
impute_dict = {
  2000:[1998,1999,2001,2002,2003,2004,
        2005,2006,2007,2008,2009],
  2010:[2011,2012,2013,2014,2015,2016,
        2017,2018,2019,2020,2021]
}

# Imputing urbanicity data for years in data without values
for key in impute_dict.keys():
  for year in impute_dict[key]:
    urb_small.loc[:,f'year_{year}'] = urb_small[f'year_{key}']

# Stacking this df
stateyear_list = list()
for year in range(1998,2022):

  # Selecting only this year's data
  _df = urb_small[[
    'State', f'year_{year}'
  ]]

  # Renaming columns for uniformity
  _df.columns = ['state','urbanicity']

  # Adding year column
  _df.loc[:,'year'] = year

  # Appending this df to list for concatenation
  stateyear_list.append(_df)

# Concatenate these lists together while dropping nulls
stateyear_urb = pd.concat(stateyear_list).dropna()

# Preview results
stateyear_urb.info()
```

```python
# Just making sure things look okay
stateyear_urb.groupby('year', as_index=False).tail(8)
```

```python
# Export to CSV
stateyear_urb.to_csv('../04 - Data/Final Data/urb_by_state-year.csv')
```
